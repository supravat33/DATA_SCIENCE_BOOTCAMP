### seaborn
---> it is used for statitical visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns#statitic visualiztion
%matplotlib inline
plt.rcParams['figure.figsize']=(6,2)
import warnings
warnings.filterwarnings('ignore')
```

```python
df['InternetUsers']
```

```python
vis1=sns.distplot(df['InternetUsers'])
plt.show()
```

```python
vis2=sns.displot(df['InternetUsers'])
plt.show()
```

```python
vis3=sns.distplot(df['InternetUsers'],bins=15)
plt.show()
```

### univarient analysis-creating the plot using single varient

```python
vis4=sns.distplot(df['InternetUsers'],bins=10)
plt.show()
```

###  outlier-stat outlier is the data point which is very far from observation
### box plot-min point,max point,average
### out lier is also called anomoly detection

```python
vis5=sns.boxplot(data=df,x='IncomeGroup',y='BirthRate',color='grey')
plt.show()
```

```python
vis6=sns.lmplot(data= df,x='InternetUsers',y='BirthRate')
plt.show()
```

```python
vis7=sns.lmplot(data= df,x='InternetUsers',y='BirthRate',fit_reg=False) #lm - linear model
plt.show()
```

```python
vis8 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup') #hue - parameter for color
plt.show()
```

```python
vis5 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup',height = 10)
plt.show()
```
