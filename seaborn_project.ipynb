# Kernal Density Estimate plot ( KDE PLOT) 

```python
# where do u find more density and how density is distibuted across from the the chat 
# center point is kernal this is calld KDE & insteade of dots it visualize like this
# we can able to clearly see the spread at the audience ratings
kd1=sns.kdeplot(x=movies.CriticRating,y=movies.AudianceRating)
```

```python
x=movies.CriticRating

k1 = sns.kdeplot(x=movies.CriticRating,y=movies.AudianceRating,shade = True,shade_lowest=False,cmap='Reds')
```

```python
k2 = sns.kdeplot(x=movies.CriticRating,y=movies.AudianceRating,shade=False,shade_lowest=False,cmap='Greens_r')
```
